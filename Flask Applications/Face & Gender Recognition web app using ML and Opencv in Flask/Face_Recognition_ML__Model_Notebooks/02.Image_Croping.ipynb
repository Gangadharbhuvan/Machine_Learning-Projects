```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
from PIL import Image 
from glob import glob
import pickle
```

```python
femalepath = glob('./data/female/*jpg')
malepath = glob('./data/male/*jpg')
```

```python
len(femalepath), len(malepath)
```

```python
# Let us consider one image
path = femalepath[999]
img = cv2.imread(path)
```

```python
plt.imshow(img)
plt.show()
```

```python
# convert img to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray, cmap='gray')
```

```python
# load haar cascade classifier
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
```

```python
faces = haar.detectMultiScale(gray,1.5,5)
print(faces)
```

```python
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
plt.imshow(img)
```

```python
# crop the image
crop_img = img[y:y+h,x:x+h]
```

```python
plt.imshow(crop_img)
```

```python
# save the image
cv2.imwrite("f_01.jpg",crop_img)
```

```python
# Apply to all the images
def extract_images(path,gender,i):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = haar.detectMultiScale(gray,1.5,5)
    for x,y,w,h in faces:
        roi = img[y:y+h,x:x+w]
        if gender == 'male':
            cv2.imwrite('./data/crop/male_crop/{}_{}.jpg'.format(gender,i),roi)
        else:
            cv2.imwrite('./data/crop/female_crop/{}_{}.jpg'.format(gender,i),roi)
```

```python
extract_images(femalepath[0],'female',1)
```

```python
for i,path in enumerate(femalepath):
    try:
        extract_images(path,'female',i)
        print("INFO: {}/{} processed Successfully".format(i,len(femalepath)))
    except:
        print("INFO: {}/{} cannot be precessed".format(i,len(femalepath)))
```

```python
for i,path in enumerate(malepath):
    try:
        extract_images(path,'male',i)
        print("INFO: {}/{} processed Successfully".format(i,len(malepath)))
    except:
        print("INFO: {}/{} cannot be precessed".format(i,len(malepath)))
```
